```python
import numpy as np
import pandas as pd
```

```python
#Temperature readings for 7 cities across 12 months
cities = ["Kampala", "Nairobi", "Tokyo", "Algiers", "Cairo", "Peru", "Lagos"]
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
temperature_data = np.array([
        [40.5, 30.5, 27.5, 45.0, 0.0, 20.6, 69.0, 58.9, 20.5, 78.5, 21.3, 40.5],
        [0.0, 50.5, 45.5, 48.5, 49.0, 42.8, 50.5, 45.6, 48.9, 45.0, 60.0, 0.0],
        [20.5, 45.8, 60.6, 54.5, 0.0, 45.8, 90.7, 85.9, 60.5, 43.3, 14.8, 90.5],
        [40.5, 50.5, 60.5, 70.0, 75.5, 55.0, 65.5, 45.0, 80.0, 30.0, 25.5, 35.0],
        [20.0, 15.5, 25.0, 30.5, 35.0, 25.5, 40.5, 45.0, 55.0, 50.5, 60.5, 65.0],
        [21.0, 32.5, 43.0, 54.5, 65.0, 65.5, 76.0, 76.6, 87.0, 98.5, 0.0,89.0],
        [12.0, 23.5, 45.0, 56.5, 67.0, 78.5, 89.0, 90.5, 0.0, 79.5, 30.6, 10.2]
    ])

    
df = pd.DataFrame(temperature_data, index = cities, columns = months)   
print("temperature readings (°C) for 7 cities over 12 months:")
print(df)

    
        
        
        
    
```

```python
#Finding the mean, meadian and standard deviation for each city
df = pd.DataFrame(temperature_data, index = cities, columns = months)
mean_values = df.mean(axis = 1)
median_values = df.median(axis = 1)
std_values = df.std(axis = 1)

summary_stats = pd.DataFrame({
    'mean(°C)': mean_values,
    'median(°C)': median_values,
    'standard deviation(°C)': std_values
})
print("\ntemperature statistics for  each city:")
print(summary_stats.round(2))
```

```python
#Finding the hottest and coldest months overall
monthly_avg = df.mean(axis = 0)
hottest_month = monthly_avg.idxmax()
coldest_month = monthly_avg.idxmin()
hottest_temp = monthly_avg.max()
coldest_temp = monthly_avg.min()

print(f"\nhottest month: {hottest_month}")
print(f"\ncoldest month: {coldest_month}")
```

```python
#Below are the displayed results in a neatly formatted table
summary_table = pd.DataFrame({
    'Month': months,
    'Average Temp(°C)': monthly_avg.round(2)
})
summary_table['Note'] = ''
summary_table.loc[summary_table['Month'] == hottest_month, 'Note'] = 'hottest'
summary_table.loc[summary_table['Month'] == coldest_month, 'Note'] = 'coldest'

print("\nMonthly average temperatures across all cities:")
print(summary_table.to_string(index = False))
```
